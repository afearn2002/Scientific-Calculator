A digital calculator is a type of electronic calculator, usually but not always handheld, designed to calculate problems in science, engineering, and mathematics. They have completely replaced slide rules in traditional applications and are widely used in both education and professional settings.

Build a Scientific calculator which will have at least 15 scientific operations (excluding the basic operations). You can use basic statistics, math module, random module.

```python
import math

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Cannot divide by zero"
    return x / y

def log10(x):
    return math.log10(x)

def square_root(x):
    return math.sqrt(x)

def power_of_two(x):
    return x ** 2

def sine(x):
    return math.sin(x)

def cosine(x):
    return math.cos(x)

def tangent(x):
    return math.tan(x)

def logarithm(x):
    return math.log(x)

def ceiling(x):
    return math.ceil(x)

def floor(x):
    return math.floor(x)

def arcsine(x):
    if -1 <= x <= 1:
        return math.asin(x)
    else:
        return "Input out of range for arcsine"

def arccosine(x):
    if -1 <= x <= 1:
        return math.acos(x)
    else:
        return "Input out of range for arccosine"

def hyperbolic_cosine(x):
    return math.cosh(x)

def log_base_2(x):
    if x <= 0:
        return "Input must be greater than 0"
    return math.log2(x)

def scientific_calculator():
    print("Scientific Calculator")
    print("Available Operations:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Log base 10")
    print("6. Square Root")
    print("7. Power of 2")
    print("8. Sine")
    print("9. Cosine")
    print("10. Tangent")
    print("11. Natural Logarithm")
    print("12. Ceiling")
    print("13. Floor")
    print("14. Arcsine")
    print("15. Arccosine")
    print("16. Hyperbolic Cosine")
    print("17. Log base 2")
    print("0. Exit")

    while True:
        choice = int(input("Enter operation number: "))
        if choice == 0:
            print("Exiting Calculator.")
            break
        elif choice >= 1 and choice <= 17:
            if choice in [1, 2, 3, 4]:
                x = float(input("Enter first number: "))
                y = float(input("Enter second number: "))
            else:
                x = float(input("Enter the number: "))

            if choice == 1:
                print("Result:", add(x, y))
            elif choice == 2:
                print("Result:", subtract(x, y))
            elif choice == 3:
                print("Result:", multiply(x, y))
            elif choice == 4:
                print("Result:", divide(x, y))
            elif choice == 5:
                print("Result:", log10(x))
            elif choice == 6:
                print("Result:", square_root(x))
            elif choice == 7:
                print("Result:", power_of_two(x))
            elif choice == 8:
                print("Result:", sine(x))
            elif choice == 9:
                print("Result:", cosine(x))
            elif choice == 10:
                print("Result:", tangent(x))
            elif choice == 11:
                print("Result:", logarithm(x))
            elif choice == 12:
                print("Result:", ceiling(x))
            elif choice == 13:
                print("Result:", floor(x))
            elif choice == 14:
                print("Result:", arcsine(x))
            elif choice == 15:
                print("Result:", arccosine(x))
            elif choice == 16:
                print("Result:", hyperbolic_cosine(x))
            elif choice == 17:
                print("Result:", log_base_2(x))
        else:
            print("Invalid input. Please enter a valid operation number.")

if __name__ == "__main__":
    scientific_calculator()
```
